##### CONFIGURACION INICIAL

```python
# LIBRER√çAS USADAS
import pandas as pd
```

```python
# DESCARGAR DATOS
airports_data = pd.read_csv('https://ourairports.com/data/airports.csv')
airports_data.head()
```

##### TRANSFORMACIONES

```python
# REMOVE COLUMNS
airports_data2 = airports_data[["iata_code", "name", "latitude_deg", "longitude_deg", "municipality", "iso_country", "continent"]]
```

```python
# NOS QUEDAMOS CON LOS AEROPUERTOS DE EUROPA, ARABIA SAUDI Y AUSTRALIA
# DE EUROPA DESCARTAMOS AQUELLOS QUE ESTEN MAS AL ESTE DE MOSCU
# DESCARTAMOS LOS AEROPUERTOS SIN CODIGO IATA YA QUE SON OTRAS COSAS COMO HELIPUERTOS O SON AEROPUERTOS QUE NO SON DE INTERES
airports_data3 = airports_data2.loc[airports_data2.iata_code.notnull()]
airports_data3 = airports_data3.loc[airports_data3.iata_code != '0']
airports_data3 = airports_data3.loc[((airports_data3.continent == "EU") & (airports_data3.longitude_deg.between(-25, 40)) & (airports_data3.latitude_deg.between(25, 75))) | (airports_data3.iso_country.isin(["SA", "AU"]))]
# UN AEROPUERTO EN EUROPA TIENE UN REGISTRO DUPLICADO
airports_data3.drop_duplicates(inplace = True)
airports_data3.head(10)
```

```python
# RENOMBRAR COLUMNAS
column_names = {'latitude_deg':'airport_latitude',
    "longitude_deg":"airport_longitude",
    "iata_code":"airport_id",
    "name":"airport_name",
    "iso_country": "airport_country_iso",
    "continent": "airport_continent_iso",
    "municipality": "airport_city"
    }
airports_data3.rename(columns = column_names, inplace = True)
airports_data3.head(10)
```

```python
airports_data3.to_csv(r'..\data_out\airports_F.csv', sep=";", decimal=',', index = False)
```
