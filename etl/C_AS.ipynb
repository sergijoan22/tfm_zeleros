```python
# LIBRERIAS USADAS
import pandas as pd
from datetime import datetime
```

```python
# LEER CSV
cities_world = pd.read_csv(r'..\data_in\C\W\worldcities.csv', sep=",", decimal = '.')
cities_world.head()
```

```python
# FILTER TO GET ONLY CITIES FROM SAUDI ARABIA
cities_AS = cities_world.loc[cities_world['country'] == 'Saudi Arabia']
cities_AS.head()
```

```python
cities_AS = cities_AS[["city_ascii", "lat", "lng", "population"]]
```

```python
# columna con el nombre de la ciudad y el pais
cities_AS['full_name'] = cities_AS.apply(lambda x: x['city_ascii'] + ', ' + 'SA', axis=1)
```

```python
cities_AS.head()
```

```python
# RENAME COLUMNS
column_names = {'city_ascii':'name',
    "lat":"latitud",
    "lng":"longitude"}
cities_AS.rename(columns = column_names, inplace = True)
```

```python
cities_AS.head()
```

```python
cities_AS.to_csv(r'..\data_out\cities_AS.csv', sep=";", decimal=',',  index = False)
```
